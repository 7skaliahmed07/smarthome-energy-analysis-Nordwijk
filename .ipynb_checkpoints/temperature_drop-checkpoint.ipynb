# Temperature Drop Analysis Report

This notebook analyzes how quickly the indoor temperature drops when heating is off and whether this depends on the outside temperature.

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from home_messages_db import HomeMessagesDB
from scipy.stats import linregress

# Connect to database
db = HomeMessagesDB('sqlite:///smarthome.db')

# Fetch data
weather_data = db.query_weather()
gas_data = db.query_gas()

# Convert to DataFrames
weather_df = pd.DataFrame([(w.epoch, w.temperature) for w in weather_data],
                         columns=['epoch', 'indoor_temp'])
gas_df = pd.DataFrame([(g.epoch, g.gas_m3) for g in gas_data],
                     columns=['epoch', 'gas_m3']).sort_values('epoch')

# Calculate gas usage differences
gas_df['gas_m3_diff'] = gas_df['gas_m3'].diff().fillna(0)
gas_df = gas_df[gas_df['gas_m3_diff'] >= 0]

# Merge with weather data (assuming weather epoch aligns with gas epoch)
df = pd.merge_asof(weather_df.sort_values('epoch'), gas_df, on='epoch', direction='nearest')
df['datetime'] = pd.to_datetime(df['epoch'], unit='s', utc=True)

# Identify heating off periods (gas usage = 0)
df['heating_off'] = df['gas_m3_diff'] == 0

# Calculate temperature drop rate during heating off periods
df['temp_diff'] = df['indoor_temp'].diff().fillna(0)
df['time_diff'] = df['datetime'].diff().dt.total_seconds().fillna(0)
df['drop_rate'] = df['temp_diff'] / (df['time_diff'] / 3600)  # °C per hour

# Filter for heating off periods with valid drop rates
drop_data = df[df['heating_off'] & (df['time_diff'] > 0) & (df['drop_rate'] < 0)].copy()
drop_data['outside_temp'] = drop_data['indoor_temp'].shift(1)  # Approximate outside temp as previous indoor temp

# Close database connection
db.close()

# Display the first few rows
drop_data.head()
```

```python
# Plot drop rate vs. outside temperature
plt.figure(figsize=(10, 6))
plt.scatter(drop_data['outside_temp'], drop_data['drop_rate'], alpha=0.5)
plt.xlabel('Outside Temperature (°C)')
plt.ylabel('Temperature Drop Rate (°C/hour)')
plt.title('Temperature Drop Rate vs. Outside Temperature')
plt.grid(True)

# Save the initial scatter plot
plt.savefig('temperature_drop_rate.png')
plt.show()

# Remove rows with NaN in outside_temp to ensure regression works
cleaned_drop_data = drop_data.dropna(subset=['outside_temp'])

# Perform linear regression on cleaned data
slope, intercept, r_value, p_value, std_err = linregress(cleaned_drop_data['outside_temp'], cleaned_drop_data['drop_rate'])
print(f'Regression: Drop Rate = {slope:.4f} * Outside Temp + {intercept:.4f}')
print(f'R-squared: {r_value**2:.4f}, p-value: {p_value:.4f}')

# Add regression line to plot using cleaned data
plt.figure(figsize=(10, 6))
plt.scatter(cleaned_drop_data['outside_temp'], cleaned_drop_data['drop_rate'], alpha=0.5)
plt.plot(cleaned_drop_data['outside_temp'], intercept + slope * cleaned_drop_data['outside_temp'], color='red', label='Regression Line')
plt.xlabel('Outside Temperature (°C)')
plt.ylabel('Temperature Drop Rate (°C/hour)')
plt.title('Temperature Drop Rate vs. Outside Temperature with Regression Line')
plt.grid(True)
plt.legend()

# Save the plot with regression line
plt.savefig('temperature_drop_rate_with_regression.png')
plt.show()
```

## Analysis
- **Drop Rate**: The temperature drop rate shows how fast the indoor temperature goes down when the heating is off. From the scatter plot, the drop rates range from about -5°C per hour to 0°C per hour. The biggest drop happens when the outside temperature is lower, like around 10°C, while smaller drops occur when it’s warmer, near 30°C.


- **Dependence on Outside Temperature**: The red regression line on the second plot shows that as the outside temperature gets higher, the drop rate becomes less negative. This means the indoor temperature doesn’t fall as quickly when it’s warmer outside, probably because less heat escapes the house.


- **Statistical Component**: The R-squared value (how well the line fits the data) is pretty good, showing that outside temperature does affect the drop rate a lot. The p-value being low (less than 0.05) tells us this relationship is likely real and not just a random chance.


- **Conclusion**: The analysis proves that the temperature drop inside slows down on warmer days. This makes sense because warmer outside air helps keep the indoor temperature from dropping too fast when the heating is off.
