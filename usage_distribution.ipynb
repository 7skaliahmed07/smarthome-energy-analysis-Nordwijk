# Hourly Usage Distribution Report

This notebook visualizes the distribution of electricity and gas usage over a day, based on the aggregated data in `hourly_usage.csv`. The data represents the average hourly usage across the period from 2022-10-09 to 2025-01-30 (limited by SmartThings data range).

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('hourly_usage.csv')

# Display the first few rows
df.head()
```

```python
# Plot the hourly usage distribution
plt.figure(figsize=(10, 6))
plt.plot(df['hour'], df['t1_kwh'], label='T1 Electricity (kWh)', marker='o')
plt.plot(df['hour'], df['t2_kwh'], label='T2 Electricity (kWh)', marker='o')
plt.plot(df['hour'], df['gas_m3'], label='Gas (m)', marker='o')
plt.xlabel('Hour of Day')
plt.ylabel('Average Usage')
plt.title('Hourly Usage Distribution (Electricity and Gas)')
plt.legend()
plt.grid(True)

# Save the plot
plt.savefig('hourly_usage_plot.png')
plt.show()
```

## Analysis of Gas/Electricity : 
- **Electricity Usage (T1 and T2)**:
  - T1 peaks around hour 21 (~0.123 kWh), likely due to evening activities (e.g., lighting, cooking).
  - T2 is mostly 0, with small peaks (e.g., 0.085 kWh at hour 20), indicating minimal off-peak usage.
- **Gas Usage**:
  - Peaks at hour 21 (0.0544 m), possibly due to heating or hot water usage in the evening.
  - Lowest in early morning (e.g., 0.0185 m at hour 2), aligning with reduced activity during sleep.
- **Patterns**:
  - Both electricity and gas usage show evening peaks, suggesting higher activity in the household during these hours.
  - Low usage in early morning hours (05) indicates minimal activity, likely when occupants are asleep.
